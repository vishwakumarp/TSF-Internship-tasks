```
#importing necessarry packages
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
pd.plotting.register_matplotlib_converters()
import seaborn as sns
```

## **Reading Data**

```
#Loading Data
data_url="http://bit.ly/w-data"
mrk_data = pd.read_csv(data_url)
```

```
print(mrk_data)
```

```
mrk_data.describe()
```

## **Visualizing Data**

```
#Visualizing using lineplot,barplot,heatmap,scatterplot respectively

plt.figure(figsize=(8,6))
sns.lineplot(data=mrk_data)
plt.figure(figsize=(8,6))
sns.barplot(x=mrk_data.Scores,y=mrk_data.Hours)
plt.figure(figsize=(8,6))
sns.heatmap(data=mrk_data,annot=True)
plt.figure(figsize=(8,6))
sns.scatterplot(x=mrk_data.Scores,y=mrk_data.Hours)
```

**It is clear that we have very little data to train our algorithm so we need to be precise on our model selection , but let's first split the data into test and train data**

```


X = mrk_data.Hours
y = mrk_data.Scores
X=X.values.reshape(-1,1)
from sklearn.model_selection import train_test_split
train_X,val_X,train_y,val_y = train_test_split(X,y,random_state=5)
```

## **Model Training**

### Decision Tree Regressor

```
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor(random_state=5)
#Decision Tree Regressor
dtr.fit(train_X,train_y)
pred_y = dtr.predict(val_X)

from sklearn.metrics import mean_absolute_error
mae1 = mean_absolute_error(val_y,pred_y)
print(mae1)

```

### Random Forest Regressor 

```
 from sklearn.ensemble import RandomForestRegressor
 rfr = RandomForestRegressor(random_state=1)
#Training Random Forest Regressor model
rfr.fit(train_X,train_y)
pred_y = rfr.predict(val_X)

mae2 = mean_absolute_error(val_y,pred_y)
print(mae2)

```

```
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

lr.fit(train_X,train_y)
pred_y = lr.predict(val_X)

mae3 = mean_absolute_error(val_y,pred_y)
print(mae3)


```

```
#Pictorial representation of Linear Regression line 
plt.figure(figsize=(8,6))
sns.regplot(x=mrk_data.Scores,y=mrk_data.Hours)
```

## **Predicting with the Given Value**

```
#predicting for given value
hrs=[[9.25]]
pred_per = lr.predict(hrs)
print("Predicted score to be obtained after studying 9.25 hours per day is",pred_per[0])
```
